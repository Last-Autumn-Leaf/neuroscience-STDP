# Partie 1 : étude des différentes formes des STDP
Pour les différentes configurations de la STDP (forme I et forme II - ex-
citatrice à excitatrice et excitatrice à inhibitrice), nous allons analysez la dynamique,
l’évolution des poids, et la stabilisation du réseau.
Nous allons prendre un neurone LIF et le connecter via des synapses à un PoissonGroup afin de voir l'évolution
des poids des synapses

```python
from tools import *
%matplotlib inline 
start_scope()


N = 100
F=15 *Hz
poisson_input = PoissonGroup(N, rates=F)

spikemon = SpikeMonitor(poisson_input)

run(50*ms)

plot(spikemon.t/ms, spikemon.i, '.k')
xlabel('Time (ms)')
ylabel('Neuron index');
```
